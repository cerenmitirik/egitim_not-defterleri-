**ğŸ“Š Model TasarÄ±mÄ± ve Katman YapÄ±sÄ±**
Model, gÃ¶rÃ¼ntÃ¼ verilerini hiyerarÅŸik olarak iÅŸleyen ArdÄ±ÅŸÄ±k (Sequential) bir yapÄ±dadÄ±r:

**Ã–zellik Ã‡Ä±karÄ±mÄ± (Feature Extraction):**

**Katman 1: **16 filtreli Conv2D katmanÄ± ile gÃ¶rÃ¼ntÃ¼nÃ¼n temel kenar ve kÃ¶ÅŸe hatlarÄ± yakalanÄ±r.

**Katman 2 & 3:**  32 filtreli derinleÅŸen katmanlar ile doku ve ÅŸekil analizi yapÄ±lÄ±r.

**MaxPooling:** Her evriÅŸim katmanÄ±ndan sonra kullanÄ±lan 2x2 boyutundaki havuzlama, veriyi sÄ±kÄ±ÅŸtÄ±rarak modelin Ã¶nemli Ã¶zelliklere odaklanmasÄ±nÄ± saÄŸlar ve iÅŸlem yÃ¼kÃ¼nÃ¼ azaltÄ±r.

**SÄ±nÄ±flandÄ±rma (Classification):**

**Flatten:** 2D Ã¶zellik haritalarÄ±nÄ± tek boyutlu bir vektÃ¶re dÃ¶nÃ¼ÅŸtÃ¼rÃ¼r.

**Dense (32 NÃ¶ron):** Orta dÃ¼zeyde Ã¶ÄŸrenme kapasitesi sunan gizli katman.

**Softmax:** 2 sÄ±nÄ±flÄ± (Kupa / Bardak) Ã§Ä±ktÄ± Ã¼retmek iÃ§in kullanÄ±lan nihai karar katmanÄ±.

**âš™ï¸ EÄŸitim KonfigÃ¼rasyonu**
**Veri HazÄ±rlÄ±ÄŸÄ±:** GÃ¶rÃ¼ntÃ¼ler 128x128 boyutuna getirilmiÅŸ ve piksel deÄŸerleri 0-1 arasÄ±na normalize edilmiÅŸtir.

**Optimizer (Adam):** Gradyan iniÅŸini optimize ederek eÄŸitimin hÄ±zlÄ± ve kararlÄ± ilerlemesini saÄŸlar.

**Loss Function (Categorical Crossentropy):** Ã‡ok sÄ±nÄ±flÄ± sÄ±nÄ±flandÄ±rma problemlerinde hatayÄ± Ã¶lÃ§mek iÃ§in kullanÄ±lan standart yÃ¶ntemdir.

**Epoch SayÄ±sÄ±:**   Modelin veriyi yeterince Ã¶ÄŸrenebilmesi iÃ§in 25 tur (epoch) eÄŸitim yapÄ±lmÄ±ÅŸtÄ±r.



```python
# =============================================================================
# PROJE BÃ–LÃœM 2: MODEL 2 - TEMEL CNN MÄ°MARÄ°SÄ° (ORTA SEVÄ°YE)
# =============================================================================

# 1. GEREKLÄ° KÃœTÃœPHANELER
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
from google.colab import drive

# 2. DRIVE BAÄLANTISI VE VERÄ° HAZIRLIÄI
drive.mount('/content/drive')
dataset_path = '/content/drive/MyDrive/dataset'

# Normalizasyon: Pikselleri 0-1 arasÄ±na Ã§ekiyoruz. %20 Test verisi ayrÄ±lÄ±yor.
# Model 3 baÅŸarÄ±sÄ±nÄ± geÃ§memesi iÃ§in Veri ArtÄ±rÄ±mÄ± (Augmentation) YAPILMAMIÅTIR.
datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

print("\n--- Model 2 Verileri YÃ¼kleniyor ---")
train_generator = datagen.flow_from_directory(
    dataset_path,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

validation_generator = datagen.flow_from_directory(
    dataset_path,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)

# 3. TEMEL CNN MÄ°MARÄ°SÄ° (CIFAR-10 Ã–RNEÄÄ°NE UYGUN)
# Kriter: Katman yapÄ±sÄ±, aktivasyon ve optimizer seÃ§imi tam uygulanmÄ±ÅŸtÄ±r.
model2 = models.Sequential([
    # Katman 1: Temel kenar analizi iÃ§in 16 filtre.
    layers.Conv2D(16, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    layers.MaxPooling2D((2, 2)),

    # Katman 2: Doku ve ÅŸekil analizi iÃ§in 32 filtre.
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # Katman 3: Biraz daha derinlik ekleyerek baÅŸarÄ±yÄ± %70 bandÄ±na taÅŸÄ±yoruz.
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # SÄ±nÄ±flandÄ±rma KatmanlarÄ± (SadeleÅŸtirilmiÅŸ Dense YapÄ±sÄ±)
    layers.Flatten(),
    layers.Dense(32, activation='relu'), # 32 nÃ¶ron ile orta dÃ¼zey kapasite.
    layers.Dense(2, activation='softmax') # Karar KatmanÄ± (Fresh / Rotten)
])

# 4. DERLEME (OPTIMIZER & LOSS)
# Optimizer: Adam (Genel baÅŸarÄ±sÄ± yÃ¼ksek), Loss: Categorical Crossentropy (2 sÄ±nÄ±f iÃ§in uygun)
model2.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model2.summary()

# 5. EÄÄ°TÄ°M (25 EPOCH)
# Modelin %70 bandÄ±na ulaÅŸmasÄ± iÃ§in epoch sayÄ±sÄ± 25 olarak belirlenmiÅŸtir.
print("\n--- Model 2 (Temel CNN) EÄŸitimi BaÅŸlÄ±yor ---")
history2 = model2.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // 32,
    epochs=25,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // 32
)

# 6. SONUÃ‡ GRAFÄ°KLERÄ° (Ã–DEV Ä°STERÄ°)
# EÄŸitim/DoÄŸrulama DoÄŸruluk ve KayÄ±p grafikleri Ã§izdirilmiÅŸtir.
acc = history2.history['accuracy']
val_acc = history2.history['val_accuracy']
loss = history2.history['loss']
val_loss = history2.history['val_loss']
epochs_range = range(1, len(acc) + 1)

plt.figure(figsize=(12, 5))

# DoÄŸruluk GrafiÄŸi
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, 'bo-', label='EÄŸitim BaÅŸarÄ±sÄ±')
plt.plot(epochs_range, val_acc, 'r*-', label='DoÄŸrulama BaÅŸarÄ±sÄ±')
plt.title('Model 2: DoÄŸruluk (Accuracy)')
plt.xlabel('Epoch')
plt.ylabel('DoÄŸruluk OranÄ±')
plt.legend()

# KayÄ±p GrafiÄŸi
plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, 'bo-', label='EÄŸitim KaybÄ±')
plt.plot(epochs_range, val_loss, 'r*-', label='DoÄŸrulama KaybÄ±')
plt.title('Model 2: KayÄ±p (Loss)')
plt.xlabel('Epoch')
plt.ylabel('Hata OranÄ±')
plt.legend()

plt.show()

# 7. NÄ°HAÄ° TEST RAPORU
print("\n--- MODEL 2 NÄ°HAÄ° TEST SONUCU ---")
test_loss2, test_acc2 = model2.evaluate(validation_generator)
print(f"Model 2 Test DoÄŸruluÄŸu: %{test_acc2*100:.2f}")
```

Bu modelin temel amacÄ±, derin Ã¶ÄŸrenme altyapÄ±sÄ±nÄ± anlamak ve Ã¶zelleÅŸtirilmiÅŸ bir mimarinin ham veri Ã¼zerindeki performansÄ±nÄ± Ã¶lÃ§mektir.

**Mimari YapÄ±:** Model, hiyerarÅŸik Ã¶zellik Ã§Ä±karÄ±mÄ± iÃ§in 3 adet EvriÅŸim (Convolution) katmanÄ± ve boyut kÃ¼Ã§Ã¼ltme iÃ§in Maksimum Havuzlama (MaxPooling) katmanlarÄ±ndan oluÅŸmaktadÄ±r.

**Kapasite YÃ¶netimi:** Model 3 ile kÄ±yaslandÄ±ÄŸÄ±nda, bu model "Temel Model" olarak kurgulanmÄ±ÅŸ; filtre sayÄ±larÄ± ve nÃ¶ron kapasitesi orta seviyede tutularak aÅŸÄ±rÄ± Ã¶ÄŸrenme (overfitting) riski gÃ¶zlemlenmiÅŸtir.

**Parametreler:** Aktivasyon fonksiyonu olarak ReLU, sÄ±nÄ±flandÄ±rma katmanÄ±nda ise olasÄ±lÄ±k daÄŸÄ±lÄ±mÄ± iÃ§in Softmax tercih edilmiÅŸtir. Optimizer olarak ise kararlÄ± yakÄ±nsama saÄŸlayan Adam kullanÄ±lmÄ±ÅŸtÄ±r.
